````python
# a. Find the highest absolute value at odd indexes


def highest_at_odd_index(lst):
    if not lst:  # If the list is empty
        return -1

    max_index = -1
    max_value = float("-inf")

    for i in range(1, len(lst), 2):  # Iterate over odd indexes
        if abs(lst[i]) > max_value:
            max_value = abs(lst[i])
            max_index = i

    return max_index


numbers = [4, -7, 2, -15, 8, 1]
print(highest_at_odd_index(numbers))  # Output: 3
````

````python
# b. Sum of n terms in the series S(n) = 1 + 3 + 9 + 27 + ...


def series_sum(n):
    total = 0
    term = 1  # First term is 1
    for i in range(n):
        total += term
        term *= 3  # Multiply by 3 for the next term
    return total


n = 5
print(series_sum(n))  # Output: 121 (1 + 3 + 9 + 27 + 81)
````

````python
# a. Dictionary of fruits with price >= 100


def filter_fruits_by_price(fruits, prices):
    result = {}
    for i in range(len(fruits)):
        if prices[i] >= 100:
            result[fruits[i]] = prices[i]
    return result


fruits = ["Apple", "Banana", "Cherry", "Mango"]
prices = [120, 60, 150, 90]
print(filter_fruits_by_price(fruits, prices))  # Output: {'Apple': 120, 'Cherry': 150}
````

````python
# b. Append words to "Output.txt" from a file


def write_words_to_output_file(input_filename):
    try:
        with open(input_filename, "r") as infile:  # Read from input file
            words = infile.read().split()

        with open("Output.txt", "a") as outfile:  # Append to Output.txt
            for word in words:
                outfile.write(word + "\n")
        print("Words successfully written to Output.txt")
    except FileNotFoundError:
        print(f"Error: File '{input_filename}' not found.")


# Example usage:
write_words_to_output_file("Input.txt")
````

### a. Inheritance Relationships among Classes

To structure the classes `Animal`, `Human`, `Cricketer`, `Bowler`, `Batter`, and `AllRounder` with inheritance, we can organize them in a hierarchy:

1. **Animal** is the **base class** (most generic).
2. **Human** inherits from Animal (single inheritance).
3. **Cricketer** inherits from Human (single inheritance).
4. **Bowler** and **Batter** inherit from Cricketer (single inheritance).
5. **AllRounder** inherits from both Bowler and Batter (multiple inheritance).

### Types of Inheritance Used:
1. **Single Inheritance**:  
   - `Human` inherits from `Animal`.  
   - `Cricketer` inherits from `Human`.  
   - `Bowler` and `Batter` inherit from `Cricketer`.

2. **Multiple Inheritance**:  
   - `AllRounder` inherits from both `Bowler` and `Batter`.

The class hierarchy can be visualized as:

```
       Animal
         |
       Human
         |
      Cricketer
       /     \
   Bowler    Batter
       \       /
      AllRounder
```

### b. Method Overloading and Overriding in Python

Python does not directly support method **overloading** (same method name with different parameters), but we can mimic it using default arguments or `*args`.

Method **overriding** occurs when a subclass provides a specific implementation of a method that is already defined in its parent class.

---

#### Code Implementation:

````python
# Base class
class Animal:
    def speak(self):
        print("Animals make sounds")


# Single Inheritance: Human inherits from Animal
class Human(Animal):
    def speak(self):  # Method Overriding
        print("Humans can speak")


# Cricketer inherits from Human
class Cricketer(Human):
    def play(self):  # Method specific to Cricketer
        print("Cricketer plays cricket")

    # Method Overloading (mimicked using default argument)
    def introduce(self, name=None, role=None):
        if name and role:
            print(f"I am {name}, and I play as a {role}.")
        elif name:
            print(f"I am {name}.")
        else:
            print("I am a cricketer.")


# Bowler inherits from Cricketer
class Bowler(Cricketer):
    def bowl(self):  # Method specific to Bowler
        print("Bowler bowls fast or spin")


# Batter inherits from Cricketer
class Batter(Cricketer):
    def bat(self):  # Method specific to Batter
        print("Batter focuses on batting")


# Multiple Inheritance: AllRounder inherits from both Bowler and Batter
class AllRounder(Bowler, Batter):
    def play(self):  # Method Overriding
        print("AllRounder can bowl and bat")


# Animal instance
animal = Animal()
animal.speak()  # Output: Animals make sounds

# Human instance
human = Human()
human.speak()  # Output: Humans can speak

# Cricketer instance
cricketer = Cricketer()
cricketer.speak()  # Overridden from Human: Output: Humans can speak
cricketer.play()  # Output: Cricketer plays cricket
cricketer.introduce("Rahul", "Batter")  # Overloading example
cricketer.introduce("Rahul")  # Default argument handling

# Bowler instance
bowler = Bowler()
bowler.bowl()  # Output: Bowler bowls fast or spin
bowler.introduce("Mitchell", "Bowler")  # Inherited method with overloading

# Batter instance
batter = Batter()
batter.bat()  # Output: Batter focuses on batting

# AllRounder instance
all_rounder = AllRounder()
all_rounder.play()  # Overridden: Output: AllRounder can bowl and bat
all_rounder.bowl()  # Inherited from Bowler: Output: Bowler bowls fast or spin
all_rounder.bat()  # Inherited from Batter: Output: Batter focuses on batting
````

### Explanation:

1. **Method Overriding**:  
   - `speak` method in the `Human` class overrides the `speak` method in the `Animal` class.
   - `play` method in `AllRounder` overrides the `play` method in the `Cricketer` class.

2. **Method Overloading**:  
   - The `introduce` method in the `Cricketer` class mimics overloading by using default arguments and conditional checks. It behaves differently depending on the number of arguments provided.

3. **Multiple Inheritance**:  
   - The `AllRounder` class inherits from both `Bowler` and `Batter`, demonstrating multiple inheritance. The method resolution order (MRO) ensures that methods are called from the correct base class.

### **b. How Python Interpreter Works Differently than a Compiler**

The Python interpreter and a compiler operate differently when executing code:

| **Aspect**             | **Python Interpreter**                               | **Compiler**                                   |
|------------------------|------------------------------------------------------|-----------------------------------------------|
| **Execution Process**  | Executes code line-by-line (interprets each line).   | Translates the entire code to machine code first. |
| **Speed**              | Slower due to on-the-fly interpretation.             | Faster because compiled code runs directly.   |
| **Errors**             | Stops at the first error during execution.           | Shows all errors during compilation.          |
| **Output**             | Produces output as soon as a line is executed.       | Produces an executable file after compilation.|
| **Portability**        | Platform-independent; needs the Python interpreter. | Platform-dependent; requires recompilation.   |
| **Examples**           | Python, JavaScript.                                  | C, C++, Java (after compilation to bytecode). |

### **a. MVC Pattern and its Implementation in Django**
The **Model-View-Controller (MVC)** pattern is a software design paradigm that separates an application into three interconnected components:

1. **Model**: Manages the data and business logic of the application. It directly handles the database operations.
2. **View**: Responsible for presenting data to the user. It defines the structure/layout of the interface and renders the information.
3. **Controller**: Acts as a bridge between the Model and the View. It handles user input, interacts with the Model, and determines the View to display.

#### How Django Follows MVC
Django uses a slightly modified version of MVC, often referred to as **MTV (Model-Template-View)**, but the essence of MVC remains:

1. **Model**:
   - Defined using Django's ORM (Object-Relational Mapping).
   - Represents the database structure and handles data access.
   - Example: `models.py`.

2. **View** (Controller in MVC terms):
   - These are Python functions or classes that process user requests and return a response.
   - They handle the business logic, interact with the Model, and pass data to the Template.
   - Example: `views.py`.

3. **Template** (View in MVC terms):
   - Contains the presentation layer with HTML and optional template tags.
   - Used to display data passed by the View to the end-user.
   - Example: `.html` files in the `templates` directory.

**Key Difference**: 
Django combines the roles of the Controller (business logic) and the Router (URL-to-view mapping) into the **View** component.

### **b. Comparison Between an Interpreter and a Compiler**

| Aspect                     | Interpreter                               | Compiler                                |
|----------------------------|-------------------------------------------|-----------------------------------------|
| **Execution**              | Executes code line-by-line.               | Translates the entire code into machine code before execution. |
| **Speed**                  | Slower because translation happens during runtime. | Faster after compilation as the code is pre-translated. |
| **Error Detection**        | Detects errors line-by-line during runtime. | Detects errors before execution, during compilation. |
| **Portability**            | Code is platform-independent.             | Compiled code is platform-specific.    |

#### **Advantage of Interpreter**:
- **Immediate Execution**: Since interpreters execute code line-by-line, you can test and debug scripts interactively without needing to recompile the entire program.

#### **Disadvantage of Interpreter**:
- **Performance**: Interpreters are slower because they translate code during execution, unlike compiled programs that execute pre-generated machine code.
