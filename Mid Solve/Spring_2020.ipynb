````python
import random


class BankAccount:
    def __init__(
        self, first_name, last_name, dob, nid, contact, address, initial_deposit
    ):
        # Account creation with minimum deposit check
        if initial_deposit < 1000:
            raise ValueError(
                "Minimum deposit of 1000 taka required to create an account."
            )

        self.first_name = first_name
        self.last_name = last_name
        self.dob = dob
        self.nid = nid
        self.contact = contact
        self.address = address
        self.balance = initial_deposit

        # Generate a random account number
        self.account_number = self.generate_account_number()

    def generate_account_number(self):
        # Random 10-digit account number
        return random.randint(1000000000, 9999999999)

    def show_balance(self):
        return f"Account Number: {self.account_number}\nCurrent Balance: {self.balance} taka"

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Deposit amount must be greater than zero.")
        self.balance += amount
        return f"Deposited {amount} taka. New Balance: {self.balance} taka"

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Withdrawal amount must be greater than zero.")
        if amount > self.balance:
            raise ValueError("Insufficient balance.")
        self.balance -= amount
        return f"Withdrew {amount} taka. New Balance: {self.balance} taka"


# Example usage:
try:
    # Creating a new bank account
    user1 = BankAccount(
        first_name="Sharif",
        last_name="Yousuf",
        dob="1999-01-01",
        nid="1234567890",
        contact="0123456789",
        address="Dhaka, Bangladesh",
        initial_deposit=2000,
    )

    print(user1.show_balance())  # Display current balance

    # Deposit some money
    print(user1.deposit(500))  # Deposit 500 taka

    # Withdraw some money
    print(user1.withdraw(1000))  # Withdraw 1000 taka

    # Show balance after withdrawal
    print(user1.show_balance())

except ValueError as e:
    print(e)
````

````python
class Student:
    def __init__(self, student_id, name, department):
        self.student_id = student_id
        self.name = name
        self.department = department
        self.courses = []

    def add_course(self, course_name):
        if course_name not in self.courses:
            self.courses.append(course_name)
            print(f"Course '{course_name}' added to {self.name}'s course list.")
        else:
            print(f"{self.name} is already enrolled in {course_name}.")

    def remove_course(self, course_name):
        if course_name in self.courses:
            self.courses.remove(course_name)
            print(f"Course '{course_name}' removed from {self.name}'s course list.")
        else:
            print(f"{self.name} is not enrolled in {course_name}.")

    def get_courses(self):
        return self.courses

    def update_name(self, new_name):
        self.name = new_name
        print(f"Student name updated to {self.name}.")

    def get_student_info(self):
        return {
            "Student ID": self.student_id,
            "Name": self.name,
            "Department": self.department,
            "Courses": self.courses,
        }


student1 = Student("UAP128", "Sharif Md. Yousuf", "Computer Science and Engineering")

student1.add_course("Data Communications")
student1.add_course("System Analysis and Design")

print(student1.get_student_info())

student1.remove_course("System Analysis and Design")

print(student1.get_student_info())

student1.update_name("Sharif Yousuf")

print(student1.get_student_info())
````
